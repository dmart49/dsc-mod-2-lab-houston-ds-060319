```python
#import libraries

import sqlite3
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime
from team import Team
```

```python
#importing functions

from functions import make_sql_frame
from functions import get_results
```

```python
#retrieving team names 

sql = """
            SELECT *
            FROM Unique_Teams
            ;"""
df_Team = make_sql_frame(sql)
```

```python
#creating team objects

dict_teams={}
for index, row in df_Team.iterrows():
    team = Team(row['TeamName'], row['Unique_Team_ID'])
    dict_teams[row['TeamName']] = team
dict_teams.keys()
```

```python
#creating dataFrame with only matches from 2011 season

sql_1 = """
            SELECT m.Season, m.Date, m.AwayTeam, m.HomeTeam,
            m.FTAG as Away_Goals, m.FTHG as Home_Goals, 
            m.FTR as Match_Result 
            FROM Matches as m
            WHERE m.Season == 2011 
            ;"""
match_frame = make_sql_frame(sql_1)
```

```python
#converting dates into datetime objects

match_frame['Date'] = pd.to_datetime(match_frame['Date'])
match_frame.head()
```

```python
#adding away wins, losses, draws

get_results(dict_teams, match_frame, 'A')
get_results(dict_teams, match_frame, 'H')
get_results(dict_teams, match_frame, 'D')
```

```python
#wins and losses for histograms

wins_list = []
loss_list = []
team_list = list(dict_teams.keys())

for team in team_list:
    if dict_teams[team].gp > 0:
        loss_list.append(dict_teams[team].loss)
        wins_list.append(dict_teams[team].win)
```

```python
plt.hist(wins_list, bins = 10)
plt.show()
```

```python
plt.hist(loss_list)
plt.show()
```

```python
#print teams from 2011 matches
new_list = []
for team in team_list:
    if(dict_teams[team].gp > 0):
        dict_teams[team].team_print()
        new_list.append(dict_teams[team])
```
